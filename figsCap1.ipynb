```python
import matplotlib.pyplot as plt
import networkx as nx
```

Ejemplo de grafo no dirigido

```python
# Crear grafo
G = nx.Graph()

# Nodos y posiciones (mismo layout que en TikZ)
pos = {
    1: (0, 0),
    2: (1, -1),
    3: (0, 2),
    4: (2, -1),
    5: (1, 1.2),
    6: (3, 0),
    7: (3.5, 2)
}
G.add_nodes_from(pos.keys())

# Aristas
edges = [(1, 2), (1, 5), (2, 5), (3, 4), (5, 7)]
G.add_edges_from(edges)

# Dibujar grafo
nx.draw(G, pos,
        with_labels=True,
        node_color='skyblue',
        node_size=600,
        edge_color='gray',
        font_weight='bold')

plt.axis('off')
plt.show()
```

Ejemplo de grafo dirigido

```python
# Crear grafo dirigido
G = nx.DiGraph()

# Posiciones fijas (basadas en el TikZ original)
pos = {
    1: (0, 0),
    2: (1, -1),
    3: (0, 2),
    4: (2, -1),
    5: (1, 1.2),
    6: (3, 0),
    7: (3.5, 2)
}
G.add_nodes_from(pos.keys())

# Aristas dirigidas (sentido arbitrario, puedes ajustar según tu caso)
edges = [(1, 2), (1, 5), (2, 5), (3, 4), (5, 7)]
G.add_edges_from(edges)

# Dibujar grafo dirigido
nx.draw(G, pos,
        with_labels=True,
        node_color='skyblue',
        node_size=600,
        edge_color='gray',
        arrows=True,
        arrowsize=20,
        connectionstyle='arc3,rad=0.0',
        font_weight='bold')

plt.axis('off')
plt.show()
```

Ejemplo de grafo con pesos

```python
# Crear grafo no dirigido
G = nx.Graph()

# Posiciones estilo TikZ
pos = {
    1: (0, 0),
    2: (1, -1),
    3: (0, 2),
    4: (2, -1),
    5: (1, 1.2),
    6: (3, 0),
    7: (3.5, 2)
}
G.add_nodes_from(pos.keys())

# Aristas con pesos
edges = [
    (1, 2, 3),
    (1, 5, 2),
    (2, 5, 4),
    (3, 4, 1),
    (5, 7, 5)
]
G.add_weighted_edges_from(edges)

# Dibujar grafo base
nx.draw(G, pos,
        with_labels=True,
        node_color='skyblue',
        node_size=600,
        edge_color='gray',
        font_weight='bold')

# Dibujar todas las etiquetas excepto (3,4)
edge_labels = {(u, v): w for u, v, w in edges if (u, v) != (3, 4) and (v, u) != (3, 4)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black')

# Dibujar etiqueta de (3,4) más cerca del nodo 4 (label_pos cerca de 1.0)
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels={(3, 4): 1},
    font_color='black',
    label_pos=0.8  # Más cerca del nodo 4
)

plt.axis('off')
plt.show()
```

Grafo inducido y grafo no inducido

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 4))

# ----------- GRAFO G -----------
G = nx.Graph()
G.add_nodes_from(['a', 'b', 'c', 'd', 'e'])
edges_G = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'),
           ('a', 'e'), ('e', 'c'), ('b', 'e'), ('e', 'd')]
G.add_edges_from(edges_G)

pos_G = {'a': (0, 0), 'b': (1, 0), 'c': (1, 1), 'd': (0, 1), 'e': (0.5, 0.5)}
nx.draw(G, pos_G, ax=axes[0],
        node_color='skyblue', node_size=600, edge_color='gray', with_labels=False)
axes[0].axis('off')
axes[0].text(0.5, -0.3, r'$G$', ha='center', fontsize=20, fontweight='bold')

# ----------- GRAFO G' (subgrafo inducido) -----------
G_prime = nx.Graph()
G_prime.add_nodes_from(['a', 'b', 'c', 'd'])
edges_Gp = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a')]
G_prime.add_edges_from(edges_Gp)

pos_Gp = {'a': (0, 0), 'b': (1, 0), 'c': (1, 1), 'd': (0, 1)}
nx.draw(G_prime, pos_Gp, ax=axes[1],
        node_color='skyblue', node_size=600, edge_color='gray', with_labels=False)
axes[1].axis('off')
axes[1].text(0.5, -0.3, r"$G'$", ha='center', fontsize=20, fontweight='bold')

# ----------- GRAFO G'' (no inducido) -----------
G_pp = nx.Graph()
G_pp.add_nodes_from(['a', 'b', 'c', 'd', 'e'])
edges_Gpp = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'), ('e', 'c')]
G_pp.add_edges_from(edges_Gpp)

pos_Gpp = {'a': (0, 0), 'b': (1, 0), 'c': (1, 1), 'd': (0, 1), 'e': (0.5, 0.5)}
nx.draw(G_pp, pos_Gpp, ax=axes[2],
        node_color='skyblue', node_size=600, edge_color='gray', with_labels=False)
axes[2].axis('off')
axes[2].text(0.5, -0.3, r"$G''$", ha='center', fontsize=20, fontweight='bold')

plt.tight_layout()
plt.subplots_adjust(top=0.85)  # Ajuste del espacio superior para el título global
plt.show()
```

Grafo con dos componentes conexas

```python
# Crear el grafo
G = nx.Graph()

# Componente 1 - Azul (Ciclo con un nodo adicional conectado)
G.add_nodes_from(['a1', 'a2', 'a3', 'a4', 'a5'])
edges_componente1 = [('a1', 'a2'), ('a2', 'a3'), ('a3', 'a4'), ('a4', 'a1'), ('a3', 'a5')]
G.add_edges_from(edges_componente1)

# Componente 2 - Roja (Árbol con bifurcaciones)
G.add_nodes_from(['b1', 'b2', 'b3', 'b4', 'b5', 'b6'])
edges_componente2 = [('b1', 'b2'), ('b1', 'b3'), ('b3', 'b4'), ('b3', 'b5'), ('b2', 'b6')]
G.add_edges_from(edges_componente2)

# Posiciones de los nodos (ajustadas para que las componentes sean más cuadradas)
pos = {
    'a1': (0, 0), 'a2': (1, 0), 'a3': (1, 1), 'a4': (0, 1), 'a5': (0.5, 1.5),
    'b1': (3, 0), 'b2': (3.5, 0.8), 'b3': (2.5, 0.8), 'b4': (2.2, 1.5), 'b5': (2.8, 1.5), 'b6': (3.8, 1.5)
}

# Dibujar el grafo
plt.figure(figsize=(10, 4))
nx.draw(G, pos, with_labels=False, node_color='skyblue', edge_color='gray', node_size=600, font_size=10)

plt.axis('off')  # Quitar los ejes
plt.show()
```

Puente y vértice de corte

```python
# Crear el grafo
G = nx.Graph()

# Componente 2 - Árbol con bifurcaciones
G.add_nodes_from(['b1', 'b2', 'b3', 'b4', 'b5', 'b6'])
edges_componente2 = [('b1', 'b2'), ('b1', 'b3'), ('b3', 'b4'), ('b3', 'b5'), ('b2', 'b6')]
G.add_edges_from(edges_componente2)

# Posiciones de los nodos
pos = {
    'b1': (3, 0), 'b2': (3.5, 0.8), 'b3': (2.5, 0.8), 'b4': (2.2, 1.5), 'b5': (2.8, 1.5), 'b6': (3.8, 1.5)
}

# Dibujar el grafo
plt.figure(figsize=(8, 6))

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, edgelist=[('b1', 'b2'), ('b1', 'b3'), ('b3', 'b4'), ('b2', 'b6')],
                       edge_color='gray', width=2)
nx.draw_networkx_edges(G, pos, edgelist=[('b3', 'b5')], edge_color='orange', width=2)  # El puente en naranja

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=600)

# Resaltar el nodo b1 en rojo
nx.draw_networkx_nodes(G, pos, nodelist=['b1'], node_color='red', node_size=600)

# Quitar etiquetas de los nodos
plt.axis('off')  # Quitar los ejes
plt.show()
```

Grafo bipartito

```python
# Crear el grafo bipartito
G = nx.Graph()

# Añadir nodos para la clase roja (6 nodos)
G.add_nodes_from(['r1', 'r2', 'r3', 'r4', 'r5', 'r6'])

# Añadir nodos para la clase azul (4 nodos)
G.add_nodes_from(['b1', 'b2', 'b3', 'b4'])

# Conexiones (grafo bipartito completo K_{6,4})
edges = [('r1', 'b1'), ('r1', 'b2'), ('r1', 'b3'), ('r1', 'b4'),
         ('r2', 'b1'), ('r2', 'b2'), ('r2', 'b3'), ('r2', 'b4'),
         ('r3', 'b1'), ('r3', 'b2'), ('r3', 'b3'), ('r3', 'b4'),
         ('r4', 'b1'), ('r4', 'b2'), ('r4', 'b3'), ('r4', 'b4'),
         ('r5', 'b1'), ('r5', 'b2'), ('r5', 'b3'), ('r5', 'b4'),
         ('r6', 'b1'), ('r6', 'b2'), ('r6', 'b3'), ('r6', 'b4')]

G.add_edges_from(edges)

# Posiciones de los nodos (distribuidos aún más espaciados)
pos = {
    'r1': (0, 7), 'r2': (0, 6), 'r3': (0, 5), 'r4': (0, 4), 'r5': (0, 3), 'r6': (0, 2),
    'b1': (5, 6.5), 'b2': (5, 5), 'b3': (5, 3.5), 'b4': (5, 2)
}

# Dibujar el grafo
plt.figure(figsize=(12, 10))

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

# Dibujar los nodos de la clase roja (en rojo)
nx.draw_networkx_nodes(G, pos, nodelist=['r1', 'r2', 'r3', 'r4', 'r5', 'r6'], node_color='skyblue', node_size=600)

# Dibujar los nodos de la clase azul (en azul)
nx.draw_networkx_nodes(G, pos, nodelist=['b1', 'b2', 'b3', 'b4'], node_color='red', node_size=600)

# Quitar etiquetas de los nodos
plt.axis('off')  # Quitar los ejes
plt.show()
```
