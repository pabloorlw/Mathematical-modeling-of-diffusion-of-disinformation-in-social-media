```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

from scipy.stats import poisson
```

Distribución de Poisson vs ley de potencias

```python
# Parámetros
k = np.arange(1, 30)
lambda_poisson = 5  # parámetro para la distribución de Poisson
alpha = 2.5         # parámetro para la distribución de ley de potencias

# Distribuciones
poisson_dist = poisson.pmf(k, mu=lambda_poisson)
power_law_dist = k**(-alpha)
power_law_dist /= power_law_dist.sum()  # normalización

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(k, poisson_dist, 'o-', label='Distribución de Poisson ($\lambda=5$)')
plt.plot(k, power_law_dist, 's-', label='Distribución de ley de potencias ($\\alpha=2.5$)')
plt.xlabel('Grado $k$')
plt.ylabel('Probabilidad $P(k)$')
plt.legend()
plt.grid(True)
plt.show()
```

Graficar en escala log-log

```python
plt.figure(figsize=(10, 6))
plt.loglog(k, power_law_dist, 's-', label='Ley de Potencias ($\\alpha=2.5$)')
plt.loglog(k, poisson_dist, 'o-', label='Poisson ($\lambda=5$)')
plt.xlabel('log Grado $k$')
plt.ylabel('log Probabilidad $P(k)$')
plt.legend()
plt.grid(True, which='both')
plt.show()
```

Clique K5 vs grafo con clustering moderado

```python
# Crear una figura con dos subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# --------------------------
# Subplot 1: Clique K5
# --------------------------
G_clique = nx.complete_graph(5)  # K5 es un clique de 5 nodos
pos_clique = nx.spring_layout(G_clique, seed=42)
nx.draw(G_clique, pos_clique, with_labels=True, node_color='lightblue', ax=axes[0])

# --------------------------
# Subplot 2: Grafo con clustering
# --------------------------
G_cluster = nx.Graph()
edges = [
    (0, 1), (0, 2), (0, 3),
    (1, 2), (1, 3),
    (2, 3),
    (3, 4), (4, 5), (4, 6),
    (5, 6)
]
G_cluster.add_edges_from(edges)
pos_cluster = nx.spring_layout(G_cluster, seed=42)
nx.draw(G_cluster, pos_cluster, with_labels=True, node_color='lightgreen', ax=axes[1])

plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
